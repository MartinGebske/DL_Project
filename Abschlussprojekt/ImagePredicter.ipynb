```python
import cv2
from keras.models import load_model
```

```python
new_model = load_model("LionDuck.h5")
```

```python
cap = cv2.VideoCapture(0)


while(True):
    ret, frame = cap.read(0)
    
    cropped = cv2.resize(frame, (320,240))
    
    gray = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
    
    
    k = cv2.waitKey(1) & 0xFF
    # 'Q' f√ºr Quit
    if k == ord("q"):
        break
    
    cv2.imshow("Let me guess who`s there!",cv2.flip(gray,1))

    #print("vor reshape:", gray.shape)
    gray = gray.reshape(1,240,320,1)
    #print(gray.shape)
    guess= new_model.predict(gray)
    print(guess)
    
    #cv2.imshow("Let me guess who`s there!",cv2.flip(frame,1))

cap.release()
cv2.destroyAllWindows()
```
